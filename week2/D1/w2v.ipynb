```python
def retr(word:str)->str:
  output = glove_vectors.most_similar(word)[0]

  return (output[0], round(output[1] * 100, 2))

retr("football")
```

```python
from typing import List

query:str = "when did obama died"

def embed_query(query:str) -> List[float]:
  words = query.split()
  vectors = []
  for word in words:
    vector = glove_vectors[word]
    vectors.append(vector)

  final = sum(vectors) / len(vectors)
  return final.tolist()

query_embd = embed_query(query)
```

```python
database = [
    "serry was here yesterday",
    "obama did not die",
    "mark is a lizard",
    "girls no answer",
    "twitter is bad"
]

database_embeddings = [embed_query(data) for data in database]
database_embeddings # 5 * 25
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def searcher(query_emb, database_embds):

  return cosine_similarity([query_emb],database_embds)


cos_values = searcher(query_embd, database_embeddings)





cos_values
```

```python
from numpy import argmax

database[argmax(cos_values).item()]
```
